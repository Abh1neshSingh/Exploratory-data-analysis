```python
# Data cleaning and preparation
```

```python
# Check for missing values
missing_values = spotify_df.isnull().sum()
print("Missing values:\n", missing_values)

# Check for duplicate rows
duplicate_rows = spotify_df.duplicated().sum()
print("Duplicate rows before removal:", duplicate_rows)
```

```python
# Drop rows with missing values
spotify_df.dropna(inplace=True)

# Verify missing values are handled
missing_values_after = spotify_df.isnull().sum()
print("Missing values after removal:\n", missing_values_after)
```

```python
# Remove duplicate rows
spotify_df.drop_duplicates(inplace=True)

# Verify duplicate rows have been removed
duplicate_rows_after = spotify_df.duplicated().sum()
print("Duplicate rows after removal:", duplicate_rows_after)

# Display the cleaned data
spotify_df.head()
```

```python
# Q2.What is the distribution of popularity among the tracks in the dataset? Visualize it using a histogram
```

```python
# Set the style
sns.set(style="whitegrid", palette="deep", font_scale=1.1, rc={"figure.figsize": [12, 6]})

# Plot the distribution of popularity
plt.figure(figsize=(12, 6))
sns.histplot(spotify_df['Popularity'], bins=30, kde=True, color='blue', alpha=0.7)
plt.title('Distribution of Track Popularity', fontsize=18)
plt.xlabel('Popularity', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
```

```python
# Q3.Is there any relationship between the popularity and the duration of tracks? Explore this using a scatter plot.
```

```python
# Scatter plot of popularity vs. duration
plt.figure(figsize=(12, 6))
sns.scatterplot(data=spotify_df, x='Duration (ms)', y='Popularity', alpha=0.6)
plt.title('Popularity vs. Duration of Tracks')
plt.xlabel('Duration (ms)')
plt.ylabel('Popularity')
plt.show()
```

```python
# Q4 Which artist has the highest number of tracks in the dataset? Display the count of tracks for each artist using a countplot.
```

```python

plt.figure(figsize=(12, 6))
top_artists = spotify_df['Artist'].value_counts().head(10)
sns.countplot(data=spotify_df[spotify_df['Artist'].isin(top_artists.index)], y='Artist', order=top_artists.index, palette='viridis')
plt.title('Number of Tracks per Artist')
plt.xlabel('Count')
plt.ylabel('Artist')
plt.show()
```

```python
# Q5. What are the top 5 least popular tracks in the dataset? Provide the artist name and track name for each.
```

```python
# Top 5 least popular tracks
least_popular_tracks = spotify_df.nsmallest(5, 'Popularity')[['Artist', 'Track Name', 'Popularity']]
print("Top 5 Least Popular Tracks:\n", least_popular_tracks)
```

```python
# Q6.Among the top 5 most popular artists, which artist has the highest popularity on average?
#  Calculate and display the average popularity for each artist.
```

```python
# Calculate average popularity for each artist
average_popularity_per_artist = spotify_df.groupby('Artist')['Popularity'].mean().reset_index()

# Identify the top 5 most popular artists by average popularity
top_5_artists = average_popularity_per_artist.nlargest(5, 'Popularity')
print("Top 5 Most Popular Artists by Average Popularity:\n", top_5_artists)

# Plot average popularity of top 5 artists
plt.figure(figsize=(12, 6))
sns.barplot(data=top_5_artists, x='Popularity', y='Artist', palette='viridis')
plt.title('Average Popularity of Top 5 Most Popular Artists')
plt.xlabel('Average Popularity')
plt.ylabel('Artist')
plt.show()
```

```python
#Q7.For the top 5 most popular artists, what are their most popular tracks? List the track name for each artist.
```

```python
# Find the most popular track for each of the top 5 most popular artists
most_popular_tracks = []
for artist in top_5_artists['Artist']:
    top_track = spotify_df[spotify_df['Artist'] == artist].nlargest(1, 'Popularity')
    most_popular_tracks.append(top_track[['Artist', 'Track Name', 'Popularity']])

most_popular_tracks_df = pd.concat(most_popular_tracks)
most_popular_tracks_df
```

```python
#Q8.Visualize relationships between multiple numerical variables simultaneously using a pair plot.
```

```python


# Set style and figure size
plt.figure(figsize=(12, 8))
sns.set(style="ticks", palette="muted")

# Create pair plot
pair_plot = sns.pairplot(spotify_df[['Popularity', 'Duration (ms)']], diag_kind='hist', markers='o', height=3)

# Customizing labels and title
plt.suptitle('Pair Plot of Popularity and Duration of Tracks', fontsize=18, fontweight='bold')
plt.tight_layout()

# Adjusting tick labels and grid
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Show the plot
plt.show()
```

```python
#Q9..Does the duration of tracks vary significantly across different artists?
#    Explore this visually using a box plot or violin plot.
```

```python
# Box plot of track duration across different artists
plt.figure(figsize=(12, 6))
top_artists = spotify_df['Artist'].value_counts().head(10).index
sns.boxplot(data=spotify_df[spotify_df['Artist'].isin(top_artists)], x='Artist', y='Duration (ms)', palette='viridis')
plt.title('Variation of Track Duration Across Different Artists')
plt.xlabel('Artist')
plt.ylabel('Duration (ms)')
plt.xticks(rotation=45)
plt.show()
```

```python
#Q10..How does the distribution of track popularity vary for different artists?
#     Visualize this using a swarm plot or a violin plot.
```

```python
# Swarm plot of track popularity for different artists
plt.figure(figsize=(12, 6))
sns.swarmplot(data=spotify_df[spotify_df['Artist'].isin(top_artists)], x='Artist', y='Popularity', palette='viridis')
plt.title('Distribution of Track Popularity for Different Artists')
plt.xlabel('Artist')
plt.ylabel('Popularity')
plt.xticks(rotation=45)
plt.show()
```
