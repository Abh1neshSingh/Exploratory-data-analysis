```python
df['age'].describe()
```

```python
#----> Insights 
# Count: There are 299 records in the dataset, indicating that there are 299 heart failure patients whose ages are recorded.

# Mean: The average age of heart failure patients in the dataset is approximately 60.83 years. This suggests that the dataset is centered around this age.

# Standard Deviation: The standard deviation of approximately 11.89 indicates the spread or dispersion of ages around the mean. A higher standard deviation suggests more variability in ages among patients.

# Minimum and Maximum: The minimum recorded age is 40 years, while the maximum recorded age is 95 years. This shows the range of ages covered in the dataset.

# Quartiles (25th, 50th, and 75th percentiles):

#  25th percentile (Q1): 25% of the patients are aged 51 years or younger.
#  50th percentile (Median): 50% of the patients are aged 60 years or younger.
#  75th percentile (Q3): 75% of the patients are aged 70 years or younger.
```

```python
# Replace 
df['sex'] = df['sex'].replace({1: 'Male', 0: 'Female'})
df['high_blood_pressure'] = df['high_blood_pressure'].replace({1: 'Yes', 0: 'No'})
high_bp_percentage_by_gender = df.groupby('sex')['high_blood_pressure'].value_counts(normalize=True) * 100
print("Prevalence of High Blood Pressure:")
print(high_bp_percentage_by_gender)
```

```python
df['sex'] = df['sex'].replace({1: 'Male', 0: 'Female'})
df['high_blood_pressure'] = df['high_blood_pressure'].replace({1: 'Yes', 0: 'No'})

high_bp_percentage_by_gender = df.groupby('sex')['high_blood_pressure'].value_counts(normalize=True) * 100

plt.figure(figsize=(8, 6))
high_bp_percentage_by_gender.unstack().plot(kind='bar', color=['skyblue', 'salmon'], alpha=0.7)
plt.title('Prevalence of High Blood Pressure by Gender')
plt.xlabel('Gender')
plt.ylabel('Percentage')
plt.xticks(rotation=0)
plt.legend(title='High Blood Pressure', labels=['No', 'Yes'])
plt.grid(axis='y')
plt.show()
```
