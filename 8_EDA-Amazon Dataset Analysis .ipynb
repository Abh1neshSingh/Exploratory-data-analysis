```python
# Q3.What is the distribution of discounted prices vs. actual prices
```

```python
mean_discounted_price = amazon_df['discounted_price'].mean()
median_discounted_price = amazon_df['discounted_price'].median()
std_discounted_price = amazon_df['discounted_price'].std()

mean_actual_price = amazon_df['actual_price'].mean()
median_actual_price = amazon_df['actual_price'].median()
std_actual_price = amazon_df['actual_price'].std()

print("Discounted Price Statistics")
print(f"Mean: {mean_discounted_price}")
print(f"Median: {median_discounted_price}")
print(f"Standard Deviation: {std_discounted_price}")

print("\nActual Price Statistics")
print(f"Mean: {mean_actual_price}")
print(f"Median: {median_actual_price}")
print(f"Standard Deviation: {std_actual_price}")
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(amazon_df['discounted_price'], color='blue', label='Discounted Price', kde=True)
sns.histplot(amazon_df['actual_price'], color='red', label='Actual Price', kde=True)
plt.legend()
plt.title('Distribution of Discounted Prices vs. Actual Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Combine both prices into a single DataFrame for plotting
prices_df = pd.DataFrame({
    'Price Type': ['Discounted Price'] * len(amazon_df) + ['Actual Price'] * len(amazon_df),
    'Price': pd.concat([amazon_df['discounted_price'], amazon_df['actual_price']], ignore_index=True)
})

# Box Plot
plt.figure(figsize=(12, 6))
sns.boxplot(data=prices_df, x='Price Type', y='Price', palette='viridis')
plt.title('Box Plot of Discounted Prices vs. Actual Prices')
plt.xlabel('Price Type')
plt.ylabel('Price')
plt.show()

# Violin Plot
plt.figure(figsize=(12, 6))
sns.violinplot(data=prices_df, x='Price Type', y='Price', palette='viridis')
plt.title('Violin Plot of Discounted Prices vs. Actual Prices')
plt.xlabel('Price Type')
plt.ylabel('Price')
plt.show()
```
